```python
import findspark
findspark.init()

import pyspark
sc = pyspark.SparkContext()
```

```python
!pwd
```

```python
text_data = sc.textFile(r'/home/jovyan/my_prac/shakespeare.txt')
text_data.collect()
```

```python
words = text_data.flatMap(lambda line:line.lower().replace(',','').split(' '))
words.take(10)
```

```python
word_map = words.map(lambda word:(word,1)) 
word_count = word_map.reduceByKey(lambda a, b:a+b)
word_count.take(10)
```
